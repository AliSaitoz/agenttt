```python
import os
import torch
from google.colab import drive

# ==========================================
# 1. GEREKLÄ° KÃœTÃœPHANELERÄ° YÃœKLE
# ==========================================
# EÄŸer Ã¶nceki adÄ±mlarda kurulduysa burayÄ± atlar, deÄŸilse kurar.
try:
    import chromadb
    import transformers
except ImportError:
    !pip install -qU chromadb transformers accelerate bitsandbytes sentence-transformers langchain-chroma langchain-huggingface

# Importlar (Zincir modÃ¼lÃ¼ YOK)
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# ==========================================
# 2. AYARLAR VE GPU
# ==========================================
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

veritabani_yolu = '/content/drive/MyDrive/AgentProje/chroma_db'
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"ğŸš€ DonanÄ±m: {device.upper()} (A100 GÃ¼cÃ¼ Aktif)")

# ==========================================
# 3. TÃœRKÃ‡E MODEL (Qwen2.5) HAZIRLANIYOR
# ==========================================
print("ğŸ§  TÃ¼rkÃ§e Model (Qwen2.5) yÃ¼kleniyor...")

model_id = "Qwen/Qwen2.5-7B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True
)

# Pipeline'Ä± kuruyoruz
text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=1024,
    temperature=0.3,
    return_full_text=False
)

# ==========================================
# 4. MANUEL SOHBET DÃ–NGÃœSÃœ (Zincirsiz)
# ==========================================
if os.path.exists(veritabani_yolu):
    print("ğŸ“‚ Drive veritabanÄ± baÄŸlandÄ±.")

    # Embedding modeli (Veriyi aramak iÃ§in ÅŸart)
    embedding_func = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2",
        model_kwargs={'device': device}
    )

    # VeritabanÄ±nÄ± yÃ¼kle
    vectorstore = Chroma(
        persist_directory=veritabani_yolu,
        embedding_function=embedding_func
    )

    print("\n" + "="*50)
    print("ğŸ‡¹ğŸ‡· SÄ°STEM HAZIR! (Manuel Mod)")
    print("Hata veren 'Chain' modÃ¼lÃ¼ iptal edildi.")
    print("DoÄŸrudan veritabanÄ± ile konuÅŸuyoruz.")
    print("="*50 + "\n")

    while True:
        soru = input("\nSoru: ")
        if soru.lower() in ['q', 'exit']:
            print("GÃ¼le gÃ¼le!")
            break
        if not soru.strip(): continue

        print("â³ VeritabanÄ± taranÄ±yor ve cevap Ã¼retiliyor...")

        # ADIM 1: VeritabanÄ±ndan benzer belgeleri bul (Manuel Retrieval)
        docs = vectorstore.similarity_search(soru, k=3)

        # Bulunan belgeleri birleÅŸtir
        baglam_metni = "\n\n".join([doc.page_content for doc in docs])

        # ADIM 2: Prompt'u biz elle hazÄ±rlÄ±yoruz
        prompt = f"""AÅŸaÄŸÄ±daki "BAÄLAM" iÃ§indeki bilgileri kullanarak "SORU"yu TÃ¼rkÃ§e cevapla.

        BAÄLAM:
        {baglam_metni}

        SORU:
        {soru}

        CEVAP:"""

        # ADIM 3: Modele gÃ¶nder
        try:
            # Qwen iÃ§in chat formatÄ± daha iyidir
            messages = [
                {"role": "system", "content": "Sen yardÄ±msever bir asistansÄ±n. Sadece verilen baÄŸlama gÃ¶re TÃ¼rkÃ§e cevap ver."},
                {"role": "user", "content": prompt}
            ]

            # Formata uygun hale getir
            text_input = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

            # CevabÄ± Ã¼ret
            outputs = text_generator(text_input)
            cevap = outputs[0]["generated_text"]

            print(f"\nğŸ“¢ CEVAP:\n{cevap.strip()}")

            print("\nğŸ” Kaynaklar:")
            for doc in docs:
                print(f"- Sayfa: {doc.metadata.get('page', '?')}")
            print("-" * 50)

        except Exception as e:
            print(f"Hata: {e}")

else:
    print(f"âŒ HATA: '{veritabani_yolu}' bulunamadÄ±. Ã–nce PDF yÃ¼kleme kodunu Ã§alÄ±ÅŸtÄ±r.")
```
